## Hàm 1 – Xây dựng đồ thị từ danh sách cạnh

**Yêu cầu:** Viết hàm `build_graph(edges, directed=False)` xây dựng đồ thị từ danh sách cạnh.

- **Input:** `edges` – list các tuple `(u, v)`, `directed` – Boolean
- **Output:** Dictionary biểu diễn đồ thị
- **Cách làm:** Dùng dictionary với key là đỉnh, value là list các đỉnh kề
  - Vô hướng: thêm cả `u→v` và `v→u`
  - Có hướng: chỉ thêm `u→v`


```python
def build_graph(edges, directed=False):
    
    graph = {}
    
    for u, v in edges:
        # tạo đỉnh u nếu 
        if u not in graph:
            graph[u] = []
        # tạo đỉnh v 
        if v not in graph:
            graph[v] = []

        # Thêm cạnh u → v
        graph[u].append(v)

        # Nếu vô hướng, thêm cạnh v → u
        if not directed:
            graph[v].append(u)

    return graph
```

```python
# Test case 1: Đồ thị vô hướng
print(" Test 1: Đồ thị vô hướng")
edges1 = [
    ('A', 'B'),
    ('A', 'C'),
    ('B', 'D'),
    ('C', 'D'),
    ('D', 'E')
]

graph1 = build_graph(edges1, directed=False)
print("Đồ thị vô hướng:")
for vertex, neighbors in graph1.items():
    print(f"  {vertex}: {neighbors}")
```

```python
# Test case 2: Đồ thị có hướng
print(" Test 2: Đồ thị có hướng")
edges2 = [
    ('A', 'B'),
    ('A', 'C'),
    ('B', 'D'),
    ('C', 'D'),
    ('D', 'E')
]

graph2 = build_graph(edges2, directed=True)
print("Đồ thị có hướng:")
for vertex, neighbors in graph2.items():
    print(f"  {vertex}: {neighbors}")
```

### Giải thích
- **Khởi tạo đỉnh:** Đảm bảo mọi đỉnh đều có entry trong graph (kể cả đỉnh không có cạnh ra)
- **Vô hướng vs Có hướng:**
  - Vô hướng: mỗi cạnh tạo 2 kết nối
  - Có hướng: mỗi cạnh tạo 1 kết nối

## Hàm 2 – BFS (Breadth-First Search)

**Yêu cầu:** Viết hàm `bfs(graph, start)` duyệt đồ thị theo chiều rộng.

- **Input:** `graph` – dictionary, `start` – đỉnh bắt đầu
- **Output:** List các đỉnh theo thứ tự duyệt BFS
- Dùng Queue (FIFO) + Set visited


```python
from collections import deque

def bfs(graph, start):
    
    visited = set()
    queue = deque([start])
    visited.add(start)
    result = []

    # BFS loop
    while queue:
        
        vertex = queue.popleft()
        result.append(vertex)

        for neighbor in graph[vertex]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)

    return result
```

```python
# Test case 1
print(" Test BFS ")
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E']
}
# Test 1
result = bfs(graph, 'A')
print(f"BFS từ A: {result}")

# Test 2
result2 = bfs(graph, 'D')
print(f"BFS từ D: {result2}")
```

### Giải thích
- **Queue:** Đảm bảo duyệt theo từng lớp (đỉnh thêm trước được xử lý trước)
- **Visited:** Tránh thăm lại đỉnh (tránh vòng lặp vô hạn)
- **Thời điểm đánh dấu:** Đánh dấu visited khi thêm vào queue, không phải khi lấy ra

## Hàm 3 – DFS đệ quy (Depth-First Search)

**Yêu cầu:** Viết hàm `dfs_recursive(graph, start)` duyệt đồ thị theo chiều sâu bằng đệ quy.

- **Input:** `graph` – dictionary, `start` – đỉnh bắt đầu
- **Output:** List các đỉnh theo thứ tự duyệt DFS
- **Chiến lược:** Đệ quy + Set visited, đi sâu vào một nhánh trước khi quay lui
- **Độ phức tạp:** O(V + E), không gian O(V) cho call stack

```python
def dfs_recursive(graph, start, visited=None, result=None):
    """
    DFS - Duyệt đồ thị theo chiều sâu (đệ quy)

    Chiến lược:
    1. Đánh dấu đỉnh hiện tại là visited
    2. Thêm vào kết quả
    3. Với mỗi neighbor chưa visited, đệ quy DFS

    Input:
        graph: dictionary biểu diễn đồ thị
        start: đỉnh bắt đầu
        visited: set các đỉnh đã thăm (dùng nội bộ)
        result: list kết quả (dùng nội bộ)

    Output:
        result: list các đỉnh theo thứ tự duyệt
    """
    # tạo 
    if visited is None:
        visited = set()
    if result is None:
        result = []

    # Đánh dấu đỉnh hiện tại
    visited.add(start)
    result.append(start)

    # Đệ quy với các đỉnh kề chưa thăm
    for neighbor in graph[start]:
        if neighbor not in visited:
            dfs_recursive(graph, neighbor, visited, result)

    return result
```

```python
# Test 1
print(" Test DFS Recursive")
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E']
}

result = dfs_recursive(graph, 'A')
print(f"DFS từ A: {result}")


# Test 2
result2 = dfs_recursive(graph, 'C')
print(f"DFS từ C: {result2}")
```

### Giải thích
- **Đệ quy:** Tự động sử dụng call stack thay vì stack tường minh
- **Thứ tự:** Đi sâu vào neighbor đầu tiên trước khi xét neighbor tiếp theo
- **Visited:** Cần truyền qua các lần gọi đệ quy để tránh thăm lại
- **Độ phức tạp:** O(V + E), không gian O(V) cho call stack

## Hàm 4 – Connected Components

**Yêu cầu:** Viết hàm `count_connected_components(graph)` đếm số thành phần liên thông.

- **Input:** Dictionary biểu diễn đồ thị
- **Output:** Số lượng connected components + List các components
- **Chiến lược:** Duyệt tất cả đỉnh, với mỗi đỉnh chưa visited chạy BFS → mỗi lần BFS = 1 component
- **Độ phức tạp:** O(V + E)

```python
from collections import deque

def count_connected_components(graph):
    """
    Đếm số connected components trong đồ thị

    Chiến lược:
    1. Duyệt tất cả các đỉnh
    2. Với mỗi đỉnh chưa visited:
       - Chạy BFS từ đỉnh đó
       - Tất cả đỉnh được thăm tạo thành 1 component
    3. Số lần chạy BFS = số components

    Input:
        graph: dictionary biểu diễn đồ thị

    Output:
        count: số lượng components
        components: list các components
    """
    visited = set()
    components = []

    def bfs_component(start):
        """BFS helper để tìm một component"""
        queue = deque([start])
        visited.add(start)
        component = []

        while queue:
            vertex = queue.popleft()
            component.append(vertex)

            for neighbor in graph[vertex]:
                if neighbor not in visited:
                    visited.add(neighbor)
                    queue.append(neighbor)

        return component

    # Duyệt tất cả các đỉnh
    for vertex in graph:
        if vertex not in visited:
            # Tìm component chứa vertex
            component = bfs_component(vertex)
            components.append(component)

    return len(components), components
```

```python
# Test case 1
print(" Test Connected Components ")
print("Test 1: 1 component")
graph1 = {
    'A': ['B', 'C'],
    'B': ['A', 'D'],
    'C': ['A'],
    'D': ['B']
}

count1, comps1 = count_connected_components(graph1)
print(f"Số components: {count1}")
for i, comp in enumerate(comps1, 1):
    print(f"  Component {i}: {comp}")
```

```python
# Test case 2
print("Test 2: 3 components")
graph2 = {
    'A': ['B'],
    'B': ['A'],
    'C': ['D', 'E'],
    'D': ['C'],
    'E': ['C'],
    'F': [],
}
count2, comps2 = count_connected_components(graph2)
print(f"Số components: {count2}")
for i, comp in enumerate(comps2, 1):
    print(f"  Component {i}: {comp}")
# Output:
# Component 1: ['A', 'B']
# Component 2: ['C', 'D', 'E']
# Component 3: ['F']
```

### Giải thích
- **Tại sao đúng:** Mỗi lần BFS thăm hết một component liên thông
- **Đỉnh cô lập:** Đỉnh không có cạnh nào cũng là 1 component
- **Độ phức tạp:** O(V + E) – mỗi đỉnh và cạnh được xét đúng 1 lần
- **Ứng dụng:** Phát hiện mạng lưới con, nhóm bạn bè, phân vùng dữ liệu

# Bài 2 – Cycle Detection

## Phần A – Phát hiện chu trình trong đồ thị vô hướng

**Yêu cầu:** Viết hàm `has_cycle_undirected(graph)` phát hiện chu trình trong đồ thị vô hướng.

**Chiến lược:**
- Dùng DFS với tham số `parent`
- Nếu gặp đỉnh đã visited VÀ không phải parent → có chu trình

**Tại sao cần kiểm tra parent?**
- Đồ thị vô hướng: A-B nghĩa là có A→B và B→A
- Khi ở A đi tới B, B có cạnh về A → đây không phải chu trình, chỉ là cạnh vừa đi!

```python
def has_cycle_undirected(graph):
    """
    Phát hiện chu trình trong đồ thị vô hướng

    Chiến lược:
    - Dùng DFS với tham số parent
    - Nếu gặp đỉnh đã visited VÀ không phải parent → có chu trình

    Tại sao cần kiểm tra parent?
    - Đồ thị vô hướng: A-B nghĩa là có A→B và B→A
    - Khi đang ở A, đi tới B, sau đó B có cạnh về A
    - Đây không phải chu trình, chỉ là cạnh vừa đi!
    """
    visited = set()

    def dfs(vertex, parent):
        """
        DFS helper
        vertex: đỉnh hiện tại
        parent: đỉnh cha (đỉnh đã đi qua để tới vertex)
        """
        # Đánh dấu vertex là visited
        visited.add(vertex)

        # Duyệt các neighbors
        for neighbor in graph[vertex]:
            if neighbor not in visited:
                # Đệ quy với neighbor
                if dfs(neighbor, vertex):
                    return True
            elif neighbor != parent:
                # Gặp đỉnh đã visited và không phải cha
                # → Có chu trình!
                return True

        return False

    # Duyệt tất cả các đỉnh (đồ thị có thể không liên thông)
    for vertex in graph:
        if vertex not in visited:
            if dfs(vertex, None):
                return True

    return False
```

```python
# Test cases
print(" Test Cycle Detection - Undirected ")

# Test 1: Có chu trình
graph1 = {
    0: [1, 2],
    1: [0, 3],
    2: [0, 3],
    3: [1, 2]
}
print(f"Test 1 (có chu trình): {has_cycle_undirected(graph1)}")  # True

# Test 2: Không có chu trình 
graph2 = {
    0: [1, 2],
    1: [0, 3],
    2: [0],
    3: [1]
}
print(f"Test 2 (không có chu trình): {has_cycle_undirected(graph2)}")  # False

# Test 3: Đồ thị không liên thông với chu trình
graph3 = {
    0: [1],
    1: [0],
    2: [3, 4],
    3: [2, 4],
    4: [2, 3]
}
print(f"Test 3 (nhiều components, có chu trình): {has_cycle_undirected(graph3)}")  # True
```

### Giải thích tại sao cần kiểm tra `neighbor != parent`

Trong đồ thị vô hướng, mỗi cạnh `A-B` được biểu diễn bằng hai mục: `A→B` và `B→A`. Khi DFS đang ở đỉnh `A` và di chuyển đến `B`, thì `B` sẽ thấy `A` trong danh sách neighbor. Nếu ta chỉ kiểm tra visited mà không kiểm tra parent, ta sẽ nhầm cạnh `B→A` là một chu trình. Kiểm tra `neighbor != parent` loại bỏ trường hợp này, chỉ phát hiện chu trình thật sự.

## Phần B – Phát hiện chu trình trong đồ thị có hướng

**Yêu cầu:** Viết hàm `has_cycle_directed(graph)` phát hiện chu trình trong đồ thị có hướng.

**Three-Color Approach:**
- `WHITE (0)`: Chưa thăm
- `GRAY (1)`: Đang xử lý (trong recursion stack)
- `BLACK (2)`: Đã xử lý xong

**Quy tắc:** Nếu gặp đỉnh GRAY → có chu trình!

```python
def has_cycle_directed(graph):
    """
    Phát hiện chu trình trong đồ thị có hướng
    Sử dụng three-color approach

    WHITE: Chưa thăm
    GRAY: Đang xử lý (trong recursion stack)
    BLACK: Đã xử lý xong

    Nếu gặp đỉnh GRAY → có back edge → có chu trình!
    """
    WHITE, GRAY, BLACK = 0, 1, 2

    # Khởi tạo color cho tất cả đỉnh là WHITE
    color = {vertex: WHITE for vertex in graph}

    def dfs(vertex):
        """DFS helper với three-color"""
        # Đánh dấu đỉnh đang xử lý (GRAY)
        color[vertex] = GRAY

        # Duyệt các neighbors
        for neighbor in graph[vertex]:
            # Nếu neighbor đang GRAY → trong recursion stack → có chu trình
            if color[neighbor] == GRAY:
                return True

            # Nếu neighbor là WHITE → chưa thăm
            if color[neighbor] == WHITE:
                if dfs(neighbor):
                    return True

        # Xử lý xong đỉnh này (BLACK)
        color[vertex] = BLACK
        return False

    # Duyệt tất cả các đỉnh
    for vertex in graph:
        if color[vertex] == WHITE:
            if dfs(vertex):
                return True

    return False
```

```python
# Test cases
print(" Test Cycle Detection - Directed")

# Test 1: Có chu trình
graph1 = {
    'A': ['B'],
    'B': ['C'],
    'C': ['A']
}
print(f"Test 1 (có chu trình A→B→C→A): {has_cycle_directed(graph1)}")  # True

# Test 2: DAG (không có chu trình)
graph2 = {
    'A': ['B', 'C'],
    'B': ['D'],
    'C': ['D'],
    'D': []
}
print(f"Test 2 (DAG): {has_cycle_directed(graph2)}")  # False

# Test 3: Cross edge (không phải chu trình)
graph3 = {
    'A': ['B'],
    'B': ['C'],
    'C': [],
    'D': ['C']
}
print(f"Test 3 (cross edge, không có chu trình): {has_cycle_directed(graph3)}")  # False
```

### Giải thích tại sao cần 3 trạng thái (WHITE / GRAY / BLACK)

- Trong đồ thị **có hướng**, chỉ dùng 2 trạng thái (visited/not visited) là không đủ.
- **GRAY** đánh dấu đỉnh đang nằm trong recursion stack hiện tại → nếu gặp lại đỉnh GRAY, nghĩa là có back edge → **chu trình**.
- **BLACK** đánh dấu đỉnh đã xử lý xong hoàn toàn → gặp lại đỉnh BLACK chỉ là cross edge hoặc forward edge, **không phải** chu trình.

### Sự khác biệt với đồ thị vô hướng
- **Vô hướng:** Chỉ cần kiểm tra `parent` (2 trạng thái: visited/not visited + parent tracking)
- **Có hướng:** Cần 3 trạng thái vì có nhiều loại cạnh (back edge, cross edge, forward edge)

# Bài 3 – Topological Sort

## Phần A – Topological Sort với DFS

**Yêu cầu:** Viết hàm `topological_sort_dfs(graph)` sắp xếp topo bằng DFS.

**Chiến lược (Post-Order DFS):**
1. Kiểm tra có chu trình không (chỉ DAG mới có topo sort)
2. DFS từ mọi đỉnh chưa visited
3. Thêm đỉnh vào stack SAU KHI xử lý hết neighbors (post-order)
4. Return stack đảo ngược

```python
def topological_sort_dfs(graph):
    """
    Topological Sort sử dụng DFS (post-order)

    Chiến lược:
    1. Kiểm tra có chu trình không (chỉ DAG mới có topo sort)
    2. DFS từ mọi đỉnh chưa visited
    3. Thêm đỉnh vào stack SAU KHI xử lý hết neighbors (post-order)
    4. Return stack đảo ngược

    Return: List theo thứ tự topo, hoặc None nếu có chu trình
    """
    # Kiểm tra chu trình trước
    if has_cycle_directed(graph):
        return None

    visited = set()
    stack = []

    def dfs(vertex):
        """DFS helper với post-order"""
        # Đánh dấu visited
        visited.add(vertex)

        # Duyệt neighbors
        for neighbor in graph[vertex]:
            if neighbor not in visited:
                dfs(neighbor)

        # Thêm vào stack (post-order)
        stack.append(vertex)

    # DFS từ mọi đỉnh chưa visited
    for vertex in graph:
        if vertex not in visited:
            dfs(vertex)

    # Return stack đảo ngược
    return stack[::-1]
```

```python
# Test Topological Sort DFS
print("Test Topological Sort (DFS)")

# Test 1: DAG đơn giản
dag1 = {
    'A': ['B', 'C'],
    'B': ['D'],
    'C': ['D'],
    'D': ['E'],
    'E': []
}
result1 = topological_sort_dfs(dag1)
print(f"Test 1 - DAG: {result1}")

# Test 2: Có chu trình → None
cyclic = {
    'A': ['B'],
    'B': ['C'],
    'C': ['A']
}
result2 = topological_sort_dfs(cyclic)
print(f"Test 2 - Có chu trình: {result2}")  # None

# Test 3: DAG phức tạp hơn
dag2 = {
    5: [2, 0],
    4: [0, 1],
    2: [3],
    3: [1],
    1: [],
    0: []
}
result3 = topological_sort_dfs(dag2)
print(f"Test 3 - DAG phức tạp: {result3}")
```

## Phần B – Topological Sort với Kahn's Algorithm

**Yêu cầu:** Viết hàm `topological_sort_kahn(graph)` sử dụng BFS + in-degree.

**Kahn's Algorithm:**
1. Tính in-degree (bậc vào) của mỗi đỉnh
2. Thêm các đỉnh có in-degree = 0 vào queue
3. Lặp: Lấy đỉnh từ queue → thêm vào kết quả → giảm in-degree neighbors → nếu = 0 thêm vào queue
4. Nếu `len(result) < V` → có chu trình

```python
from collections import deque

def topological_sort_kahn(graph):
    """
    Topological Sort sử dụng Kahn's Algorithm (BFS + in-degree)

    Return: List theo thứ tự topo, hoặc None nếu có chu trình
    """
    # Bước 1: Tính in-degree cho mỗi đỉnh
    in_degree = {vertex: 0 for vertex in graph}
    for vertex in graph:
        for neighbor in graph[vertex]:
            in_degree[neighbor] = in_degree.get(neighbor, 0) + 1

    # Bước 2: Thêm đỉnh có in-degree = 0 vào queue
    queue = deque()
    for vertex in in_degree:
        if in_degree[vertex] == 0:
            queue.append(vertex)

    # Bước 3: BFS loop
    result = []
    while queue:
        vertex = queue.popleft()
        result.append(vertex)

        # Giảm in-degree của các neighbors
        for neighbor in graph[vertex]:
            in_degree[neighbor] -= 1
            # Nếu in-degree = 0 → thêm vào queue
            if in_degree[neighbor] == 0:
                queue.append(neighbor)

    # Bước 4: Kiểm tra chu trình
    if len(result) != len(graph):
        return None  # Có chu trình

    return result
```

```python
# Test Topological Sort Kahn
print("=== Test Topological Sort (Kahn's Algorithm) ===")

# Test 1: DAG đơn giản
dag1 = {
    'A': ['B', 'C'],
    'B': ['D'],
    'C': ['D'],
    'D': ['E'],
    'E': []
}
result1 = topological_sort_kahn(dag1)
print(f"Test 1 - DAG: {result1}")

# Test 2: Có chu trình → None
cyclic = {
    'A': ['B'],
    'B': ['C'],
    'C': ['A']
}
result2 = topological_sort_kahn(cyclic)
print(f"Test 2 - Có chu trình: {result2}")  # None

# Test 3: DAG phức tạp hơn
dag2 = {
    5: [2, 0],
    4: [0, 1],
    2: [3],
    3: [1],
    1: [],
    0: []
}
result3 = topological_sort_kahn(dag2)
print(f"Test 3 - DAG phức tạp: {result3}")
```

### Bài toán 1: Course Schedule

Kiểm tra có thể hoàn thành tất cả môn học không?

- **Input:** `num_courses` – số môn học (0 đến n-1), `prerequisites` – list `[[a, b]]` nghĩa là phải học b trước a
- **Output:** `True` nếu có thể học hết, `False` nếu có chu trình

```python
def can_finish(num_courses, prerequisites):
    
    # Xây dựng đồ thị
    graph = {i: [] for i in range(num_courses)}
    for a, b in prerequisites:
        graph[b].append(a)  # b phải học trước a → cạnh b → a

    # Kiểm tra chu trình bằng three-color
    return not has_cycle_directed(graph)
```

### Bài toán 2: Course Schedule II

Tìm thứ tự học hợp lệ.

- **Input:** giống Course Schedule
- **Output:** List thứ tự học, hoặc `[]` nếu không thể

```python
def find_order(num_courses, prerequisites):
    """
    Tìm thứ tự học hợp lệ

    Input: giống Course Schedule
    Output: List thứ tự học, hoặc [] nếu không thể

    Gợi ý:
    1. Xây dựng đồ thị
    2. Topological sort (Kahn's Algorithm)
    3. Có chu trình → [], không có → return order
    """
    # Xây dựng đồ thị
    graph = {i: [] for i in range(num_courses)}
    for a, b in prerequisites:
        graph[b].append(a)  # b phải học trước a → cạnh b → a

    # Topological sort bằng Kahn's Algorithm
    result = topological_sort_kahn(graph)

    # Nếu có chu trình (result is None) → return []
    if result is None:
        return []

    return result
```
