## Bài 1 – Viết 4 hàm đệ quy

Mỗi hàm phải có comment ghi rõ:
- Base case là gì
- Recursive case làm gì
- Độ phức tạp Big-O

### Hàm 1 – Tính tổng từ 1 đến n

```python
def sum_to_n(n):
    """
    Tính tổng 1 + 2 + ... + n bằng đệ quy
    
    Base case:
        - n == 0 → trả về 0 (tổng rỗng)
        - n == 1 → trả về 1 (tổng chỉ có 1 phần tử)
    
    Recursive case:
        - sum_to_n(n) = n + sum_to_n(n - 1)
        - Lấy n cộng với tổng từ 1 đến (n-1)
    
    Độ phức tạp: O(n)
        - Hàm gọi đệ quy n lần, mỗi lần thực hiện 1 phép cộng
    """
    # Base case: n = 0 hoặc n = 1
    if n == 0:
        return 0
    if n == 1:
        return 1
    
    # Recursive case: n + sum(1..n-1)
    return n + sum_to_n(n - 1)


# Test
print(sum_to_n(5))   # Kết quả: 15 (1+2+3+4+5)
print(sum_to_n(10))  # Kết quả: 55
```

```python
print(f"sum_to_n(0)= {sum_to_n(0)}")  
print(f"sum_to_n(1)= {sum_to_n(1)}") 
print(f"sum_to_n(10= {sum_to_n(100)}") 

# Kiểm tra bằng công thức n*(n+1)/2
for n in [0, 1, 5, 10, 100]:
    expected = n * (n + 1) // 2
    result = sum_to_n(n)
    status = "yes" if result == expected else "not"
    print(f"{status} sum_to_n({n}) = {result}, công thức = {expected}")
```

### Hàm 2 – Tính n mũ k (power)

```python
def power(n, k):
    """
    Tính n^k bằng đệ quy
    
    Base case:
        - k == 0 → trả về 1 (bất kỳ số nào mũ 0 đều bằng 1)
        - n == 0 → trả về 0 (0 mũ bất kỳ số dương nào đều bằng 0)
    
    Recursive case:
        - power(n, k) = n * power(n, k - 1)
        - Nhân n với n^(k-1)
    
    Độ phức tạp: O(k)
        - Hàm gọi đệ quy k lần
    
    Lưu ý: Có cách tối ưu hơn bằng 'chia để trị' (O(log k))
    nhưng không yêu cầu trong bài này.
    """
    # Base case 1: mũ 0
    if k == 0:
        return 1
    
    # Base case 2: cơ số 0
    if n == 0:
        return 0
    
    # Recursive case: n × n^(k-1)
    return n * power(n, k - 1)

print(power(2, 5))  
print(power(3, 4))  
print(power(5, 0))  
```

### Hàm 3 – Đảo chuỗi (reverse string)

```python
def reverse_string(s):
    """
    Đảo ngược chuỗi bằng đệ quy

    Base case:
        - len(s) <= 1 → trả về chính chuỗi đó
          (chuỗi rỗng hoặc 1 ký tự không cần đảo)

    Recursive case:
        - reverse_string(s) = reverse_string(s[1:]) + s[0]
        - Đảo phần còn lại rồi nối ký tự đầu vào cuối
        - Ví dụ: reverse("hello") = reverse("ello") + "h"
                                   = "olle" + "h" = "olleh"

    Độ phức tạp: O(n)
        - n lần gọi đệ quy với n = len(s)
    """
    # Base case: chuỗi rỗng hoặc 1 ký tự
    if len(s) <= 1:
        return s

    # Recursive case: đảo phần còn lại + ký tự đầu
    return reverse_string(s[1:]) + s[0]


# Test
print(reverse_string("hello"))  # Kết quả: "olleh"
print(reverse_string("python"))  # Kết quả: "nohtyp"
print(reverse_string("a"))  # Kết quả: "a"
print(reverse_string(""))  # Kết quả: ""
```

### Hàm 4 – Kiểm tra palindrome (đọc xuôi ngược như nhau)

```python
def is_palindrome(s):
    """
    Kiểm tra chuỗi có phải palindrome bằng đệ quy
    
    Base case:
        - len(s) <= 1 → trả về True
          (chuỗi rỗng hoặc 1 ký tự luôn là palindrome)
    
    Recursive case:
        - So sánh ký tự đầu s[0] và cuối s[-1]:
            + Nếu khác nhau → return False (không phải palindrome)
            + Nếu giống nhau → kiểm tra đệ quy phần giữa s[1:-1]
    
    Độ phức tạp: O(n)
        - Tối đa n/2 lần so sánh, vẫn là O(n)
    """
    # Base case: chuỗi rỗng hoặc 1 ký tự
    if len(s) <= 1:
        return True
    
    # So sánh ký tự đầu và cuối
    if s[0] != s[-1]:
        return False
    
    # Recursive case: kiểm tra phần giữa
    return is_palindrome(s[1:-1])


# Test
print(is_palindrome("racecar"))   # True
print(is_palindrome("madam"))     # True
print(is_palindrome("hello"))     # False
print(is_palindrome("a"))         # True
print(is_palindrome(""))          # True
```

## Bài 2 – Tối ưu Fibonacci với memoization

Viết 3 phiên bản Fibonacci và so sánh hiệu suất.

### Phần A – Fibonacci đệ quy đơn giản

```python
def fibonacci_naive(n):
    """
    Fibonacci đệ quy đơn giản - CHẬM
    F(n) = F(n-1) + F(n-2)
    F(0) = 0, F(1) = 1
    
    Độ phức tạp: O(2^n) - rất chậm
    """
    if n <= 1:
        return n
    return fibonacci_naive(n - 1) + fibonacci_naive(n - 2)


# Test
print("Fibonacci naive:")
print(fibonacci_naive(10))  # 55
print(fibonacci_naive(20))  # 6765 (chậm rồi)
# print(fibonacci_naive(35))  # Rất chậm! Mất vài giây
```

### Phần B – Fibonacci với memoization

```python
def fibonacci_memo(n, memo=None):
    """
    Fibonacci với memoization - NHANH
    
    Sử dụng dict memo để lưu lại các kết quả đã tính,
    tránh tính lại nhiều lần.
    
    Độ phức tạp: O(n) thời gian, O(n) không gian
    """
    # Khởi tạo memo nếu chưa có
    if memo is None:
        memo = {}
    
    # Kiểm tra n đã có trong memo chưa
    # Nếu có → trả về ngay memo[n]
    if n in memo:
        return memo[n]
    
    # Base case: n <= 1
    if n <= 1:
        memo[n] = n
        return memo[n]
    
    # Recursive case: F(n) = F(n-1) + F(n-2)
    memo[n] = fibonacci_memo(n - 1, memo) + fibonacci_memo(n - 2, memo)
    return memo[n]


# Test
print("Fibonacci memo:")
print(fibonacci_memo(10)) 
print(fibonacci_memo(50))   
print(fibonacci_memo(100))  
```

### Phần C – Fibonacci vòng lặp (iterative)

```python
def fibonacci_iterative(n):
    """
    Fibonacci dùng vòng lặp, không dùng đệ quy
    
    Dùng 2 biến prev và curr để lưu F(n-2) và F(n-1)
    Lặp từ 2 đến n, cập nhật prev và curr
    
    Độ phức tạp: O(n) thời gian, O(1) không gian
    """
    if n <= 1:
        return n
    
    prev, curr = 0, 1
    for _ in range(2, n + 1):
        prev, curr = curr, prev + curr
    
    return curr


# Test
print("Fibonacci iterative:")
print(fibonacci_iterative(10))   
print(fibonacci_iterative(50))   
print(fibonacci_iterative(100))  
```

### So sánh hiệu suất 3 phiên bản

```python
import time

print("--- So sánh hiệu suất ---\n")

# Naive: chỉ test n=30 vì 35 quá chậm
start = time.time()
result1 = fibonacci_naive(30)
time1 = time.time() - start
print(f"Naive F(30)  = {result1}, thời gian: {time1:.4f}s")

# Memo: có thể test với 100!
start = time.time()
result2 = fibonacci_memo(100)
time2 = time.time() - start
print(f"Memo F(100) = {result2}, thời gian: {time2:.6f}s")

# Iterative: nhanh nhất
start = time.time()
result3 = fibonacci_iterative(100)
time3 = time.time() - start
print(f"Iterative  F(100) = {result3}, thời gian: {time3:.6f}s")

print(f"\n Memo nhanh hơn Naive khoảng {time1 / max(time2, 1e-9):.0f} lần")
```

## Bài 3 – Cài đặt Merge Sort và Quick Sort

### Lựa chọn A – Merge Sort

```python
def merge_sort(arr):
    """
    Sắp xếp mảng bằng Merge Sort
    
    Base case:
        - len(arr) <= 1 → mảng đã sắp xếp, trả về ngay
    
    Recursive case:
        - Divide: chia đôi mảng
        - Conquer: đệ quy sắp xếp 2 nửa
        - Combine: trộn 2 nửa đã sắp xếp
    
    Độ phức tạp: O(n log n)
    """
    # Base case: mảng 0 hoặc 1 phần tử
    if len(arr) <= 1:
        return arr
    
    # Divide - chia đôi mảng
    mid = len(arr) // 2
    left = arr[:mid]
    right = arr[mid:]
    
    # Conquer - đệ quy sắp xếp 2 nửa
    left_sorted = merge_sort(left)
    right_sorted = merge_sort(right)
    
    # Combine - trộn 2 nửa đã sắp xếp
    return merge(left_sorted, right_sorted)


def merge(left, right):
    """
    Trộn 2 mảng đã sắp xếp thành 1 mảng sắp xếp
    """
    result = []
    i = j = 0
    
    # So sánh và chọn phần tử nhỏ hơn
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    
    # Thêm các phần tử còn lại
    result.extend(left[i:])
    result.extend(right[j:])
    
    return result


# Test
arr = [64, 34, 25, 12, 22, 11, 90]
sorted_arr = merge_sort(arr)
print(f"Mảng gốc:             {arr}")
print(f"Mảng sau khi sắp xếp: {sorted_arr}")
# Kết quả: [11, 12, 22, 25, 34, 64, 90]
```

### Lựa chọn B – Quick Sort

```python
def quick_sort(arr):
    """
    Sắp xếp mảng bằng Quick Sort
    
    Base case:
        - len(arr) <= 1 → mảng đã sắp xếp, trả về ngay
    
    Recursive case:
        - Chọn pivot (phần tử giữa)
        - Partition: chia thành 3 phần (< pivot, == pivot, > pivot)
        - Đệ quy sắp xếp phần trái và phải, ghép lại
    
    Độ phức tạp trung bình: O(n log n)
    Trường hợp xấu nhất: O(n^2)
    """
    # Base case
    if len(arr) <= 1:
        return arr
    
    # Chọn pivot (phần tử giữa)
    pivot = arr[len(arr) // 2]
    
    # Partition - chia thành 3 phần
    left = [x for x in arr if x < pivot]      # các phần tử < pivot
    middle = [x for x in arr if x == pivot]    # các phần tử = pivot
    right = [x for x in arr if x > pivot]      # các phần tử > pivot
    
    # Đệ quy và kết hợp
    return quick_sort(left) + middle + quick_sort(right)


# Test
arr = [64, 34, 25, 12, 22, 11, 90]
sorted_arr = quick_sort(arr)
print(f"Mảng gốc:             {arr}")
print(f"Mảng sau khi sắp xếp: {sorted_arr}")
# Kết quả: [11, 12, 22, 25, 34, 64, 90]
```

### Phần nâng cao – Đo thời gian thực thi và so sánh

```python
import time
import random

# Tạo mảng ngẫu nhiên
arr_small = [random.randint(1, 1000) for _ in range(100)]
arr_large = [random.randint(1, 10000) for _ in range(5000)]

print("=== MERGE SORT ===")

# Đo thời gian với mảng nhỏ
start = time.time()
sorted_small = merge_sort(arr_small.copy())
time_small = time.time() - start
print(f"Thời gian sắp xếp 100 phần tử:  {time_small:.6f}s")

# Đo thời gian với mảng lớn
start = time.time()
sorted_large = merge_sort(arr_large.copy())
time_large = time.time() - start
print(f"Thời gian sắp xếp 5000 phần tử: {time_large:.6f}s")

print("\n=== QUICK SORT ===")

# Đo thời gian với mảng nhỏ
start = time.time()
sorted_small_q = quick_sort(arr_small.copy())
time_small_q = time.time() - start
print(f"Thời gian sắp xếp 100 phần tử:  {time_small_q:.6f}s")

# Đo thời gian với mảng lớn
start = time.time()
sorted_large_q = quick_sort(arr_large.copy())
time_large_q = time.time() - start
print(f"Thời gian sắp xếp 5000 phần tử: {time_large_q:.6f}s")

print("\n=== PYTHON BUILT-IN SORT ===")

# So sánh với Python built-in sort
start = time.time()
sorted_builtin = sorted(arr_large)
time_builtin = time.time() - start
print(f"Thời gian Python sorted() 5000 phần tử: {time_builtin:.6f}s")

print("\n=== TỔNG KẾT ===")
print(f"Merge Sort (5000):   {time_large:.6f}s")
print(f"Quick Sort (5000):   {time_large_q:.6f}s")
print(f"Python sorted(5000): {time_builtin:.6f}s")
```
