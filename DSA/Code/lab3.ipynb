# Bài 1 – Backtracking cơ bản

Viết 4 hàm backtracking. Mỗi hàm có:
- Comment ghi rõ base case và recursive case
- In kết quả rõ ràng
- Test với ít nhất 2 trường hợp

## Hàm 1 – Tìm tất cả hoán vị (Permutations)

**Ý tưởng:** Ở mỗi vị trí, chọn một số từ các số còn lại.

**Độ phức tạp:**
- Thời gian: O(N! × N) – N! hoán vị, mỗi hoán vị tốn O(N) để copy
- Không gian: O(N) cho call stack

```python
def permutations(nums):
    """
    Tìm tất cả hoán vị của nums
    """
    result = []

    def backtrack(path, remaining):
        # Base case: Đã chọn đủ n số
        if len(path) == len(nums):
            result.append(path.copy())  # QUAN TRỌNG: phải copy!
            return

        # Recursive case: Thử từng số còn lại
        for i in range(len(remaining)):
            # CHOOSE: Chọn remaining[i]
            path.append(remaining[i])

            # EXPLORE: Đệ quy với các số còn lại (bỏ số vừa chọn)
            new_remaining = remaining[:i] + remaining[i+1:]
            backtrack(path, new_remaining)

            # UNCHOOSE: Quay lui
            path.pop()

    backtrack([], nums)
    return result


# Test
print("=== Test Permutations ===")
result1 = permutations([1, 2, 3])
print(f"Hoán vị của [1,2,3]: {result1}")
print(f"Số hoán vị: {len(result1)}")  # Kỳ vọng: 6 (= 3!)

result2 = permutations([1, 2])
print(f"\nHoán vị của [1,2]: {result2}")
print(f"Số hoán vị: {len(result2)}")  # Kỳ vọng: 2 (= 2!)
```

### Giải thích
- `path`: Lưu hoán vị đang xây dựng
- `remaining`: Các số chưa dùng
- `path.copy()`: Tạo bản sao (nếu không copy sẽ lưu reference, tất cả đều giống nhau!)
- `remaining[:i] + remaining[i+1:]`: Tạo list mới không chứa phần tử i

## Hàm 2 – Tìm tất cả tổ hợp (Combinations)

**Ý tưởng:** Chỉ chọn các số sau số vừa chọn để tránh trùng lặp. Dùng tham số `start`.

**Độ phức tạp:**
- Thời gian: O(C(n,k) × k)
- Không gian: O(k) cho call stack

```python
def combinations(nums, k):
    """
    Tìm tất cả tổ hợp k phần tử từ nums
    """
    result = []

    def backtrack(start, path):
        # Base case: Đã chọn đủ k phần tử
        if len(path) == k:
            result.append(path.copy())
            return

        # Recursive case: Thử các số từ vị trí start trở đi
        for i in range(start, len(nums)):
            # CHOOSE
            path.append(nums[i])

            # EXPLORE: Chỉ xét các số sau i (i+1)
            backtrack(i + 1, path)

            # UNCHOOSE
            path.pop()

    backtrack(0, [])
    return result


# Test
print("=== Test Combinations ===")
result1 = combinations([1, 2, 3, 4], 2)
print(f"Tổ hợp 2 từ [1,2,3,4]: {result1}")
print(f"Số tổ hợp: {len(result1)}")  # Kỳ vọng: 6 (C(4,2) = 6)

result2 = combinations([1, 2, 3], 2)
print(f"\nTổ hợp 2 từ [1,2,3]: {result2}")
print(f"Số tổ hợp: {len(result2)}")  # Kỳ vọng: 3 (C(3,2) = 3)
```

### Giải thích
- `start`: Vị trí bắt đầu trong nums
- `i + 1`: Chỉ xét các số sau vị trí i → tránh trùng lặp
- Công thức tổ hợp: C(n, k) = n! / (k! × (n-k)!)

## Hàm 3 – Tìm tất cả tập con (Subsets)

**Ý tưởng:** Giống combinations nhưng không có điều kiện k. Lưu tất cả path gặp phải.

**Độ phức tạp:**
- Thời gian: O(2^n × n) – 2^n tập con, mỗi tập tốn O(n) copy
- Không gian: O(n) cho call stack

```python
def subsets(nums):
    """
    Tìm tất cả tập con của nums
    """
    result = []

    def backtrack(start, path):
        # Base case (ngầm): Lưu tất cả tập con (không có điều kiện dừng cụ thể)
        result.append(path.copy())

        # Recursive case: Thử thêm các phần tử từ start
        for i in range(start, len(nums)):
            # CHOOSE
            path.append(nums[i])

            # EXPLORE
            backtrack(i + 1, path)

            # UNCHOOSE
            path.pop()

    backtrack(0, [])
    return result


# Test
print("=== Test Subsets ===")
result1 = subsets([1, 2, 3])
print(f"Tập con của [1,2,3]: {result1}")
print(f"Số tập con: {len(result1)}")  # Kỳ vọng: 8 (= 2^3)

result2 = subsets([1, 2])
print(f"\nTập con của [1,2]: {result2}")
print(f"Số tập con: {len(result2)}")  # Kỳ vọng: 4 (= 2^2)
```

### Giải thích
- Không có base case rõ ràng (lưu tất cả path)
- Mỗi lần gọi backtrack đều lưu path hiện tại
- Tổng số tập con: 2^n

## Hàm 4 – In tất cả chuỗi nhị phân độ dài n

**Ý tưởng:** Tương tự cây quyết định: mỗi nút có 2 nhánh (0 hoặc 1).

**Độ phức tạp:**
- Thời gian: O(2^n × n)
- Không gian: O(n)

```python
def binary_strings(n):
    """
    Tìm tất cả chuỗi nhị phân độ dài n
    """
    result = []

    def backtrack(path):
        # Base case: Đủ n ký tự
        if len(path) == n:
            result.append(''.join(path))
            return

        # Recursive case: Thử cả '0' và '1'
        for choice in ['0', '1']:
            # CHOOSE
            path.append(choice)

            # EXPLORE
            backtrack(path)

            # UNCHOOSE
            path.pop()

    backtrack([])
    return result


# Test
print("=== Test Binary Strings ===")
result1 = binary_strings(3)
print(f"Chuỗi nhị phân độ dài 3: {result1}")
print(f"Số chuỗi: {len(result1)}")  # Kỳ vọng: 8 (= 2^3)

result2 = binary_strings(2)
print(f"\nChuỗi nhị phân độ dài 2: {result2}")
print(f"Số chuỗi: {len(result2)}")  # Kỳ vọng: 4 (= 2^2)
```

### Giải thích
- Choices cố định: `['0', '1']`
- `''.join(path)`: Ghép list thành string
- Tương tự cây quyết định: mỗi nút có 2 nhánh (0 hoặc 1)

# Bài 2 – N-Queens với pruning

Cài đặt bài toán N-Queens và tối ưu bằng pruning.

## Phần A – Hàm kiểm tra hợp lệ

```python
def is_safe(board, row, col, n):
    """
    Kiểm tra đặt quân hậu ở (row, col) có hợp lệ không
    board: list lưu cột của quân hậu ở mỗi hàng
    board[i] = j nghĩa là quân hậu hàng i ở cột j
    """
    # Kiểm tra tất cả các hàng trước đó
    for prev_row in range(row):
        prev_col = board[prev_row]

        # Kiểm tra cùng cột
        if prev_col == col:
            return False

        # Kiểm tra đường chéo
        # Nếu |row1 - row2| == |col1 - col2| → cùng đường chéo
        if abs(prev_row - row) == abs(prev_col - col):
            return False

    return True


def print_board(solution, n):
    """
    In bàn cờ N×N với quân hậu
    """
    for row in range(n):
        line = ""
        for col in range(n):
            if solution[row] == col:
                line += "Q "
            else:
                line += ". "
        print(line)
    print()
```

## Phần B – N-Queens KHÔNG có pruning

Đặt quân hậu ở tất cả vị trí, chỉ kiểm tra hợp lệ khi đã đặt đủ n quân.

```python
class Counter:
    """Class để đếm số lần gọi hàm"""
    def __init__(self):
        self.total_calls = 0
        self.solutions = 0

    def report(self):
        print(f"Tổng số lần gọi: {self.total_calls}")
        print(f"Số giải pháp tìm được: {self.solutions}")


def solve_n_queens_no_pruning(n):
    """
    N-Queens KHÔNG có pruning (kiểm tra sau)
    """
    counter = Counter()
    result = []
    board = []

    def backtrack(row):
        counter.total_calls += 1

        # Base case: Đã đặt đủ n quân hậu
        if row == n:
            # Kiểm tra board có hợp lệ không bằng cách duyệt tất cả các cặp quân hậu
            valid = True
            for r1 in range(n):
                for r2 in range(r1 + 1, n):
                    # Cùng cột
                    if board[r1] == board[r2]:
                        valid = False
                        break
                    # Cùng đường chéo
                    if abs(r1 - r2) == abs(board[r1] - board[r2]):
                        valid = False
                        break
                if not valid:
                    break

            if valid:
                result.append(board.copy())
                counter.solutions += 1
            return

        # Recursive case: Thử tất cả cột (0 đến n-1)
        # KHÔNG kiểm tra is_safe trước
        for col in range(n):
            # CHOOSE
            board.append(col)

            # EXPLORE
            backtrack(row + 1)

            # UNCHOOSE
            board.pop()

    backtrack(0)
    counter.report()
    return result


# Test
print("=== Test N-Queens KHÔNG pruning (n=4) ===")
r1 = solve_n_queens_no_pruning(4)
if r1:
    print("Giải pháp đầu tiên:")
    print_board(r1[0], 4)

print("=== Test N-Queens KHÔNG pruning (n=5) ===")
r2 = solve_n_queens_no_pruning(5)
print(f"Số giải pháp: {len(r2)}")
```

## Phần C – N-Queens CÓ pruning

Dùng `is_safe()` để cắt tỉa TRƯỚC khi đệ quy.

```python
def solve_n_queens_with_pruning(n):
    """
    N-Queens CÓ pruning (kiểm tra trước)
    """
    counter = Counter()
    result = []
    board = []

    def backtrack(row):
        counter.total_calls += 1

        # Base case: Đã đặt đủ n quân hậu hợp lệ
        if row == n:
            result.append(board.copy())
            counter.solutions += 1
            return

        # Recursive case: Thử từng cột
        for col in range(n):
            # PRUNING: Kiểm tra is_safe TRƯỚC khi đệ quy
            if is_safe(board, row, col, n):
                # CHOOSE
                board.append(col)

                # EXPLORE
                backtrack(row + 1)

                # UNCHOOSE
                board.pop()

    backtrack(0)
    counter.report()
    return result


# Test
print("=== Test N-Queens CÓ pruning (n=4) ===")
r3 = solve_n_queens_with_pruning(4)
if r3:
    print("Giải pháp đầu tiên:")
    print_board(r3[0], 4)

print("=== Test N-Queens CÓ pruning (n=5) ===")
r4 = solve_n_queens_with_pruning(5)
print(f"Số giải pháp: {len(r4)}")
```

## Phần D – So sánh hiệu suất

```python
import time

def compare_n_queens(n):
    print(f"\n{'='*50}")
    print(f"So sánh N-Queens với N={n}")
    print(f"{'='*50}")

    # Test không có pruning
    print("\n[1] KHÔNG có pruning:")
    start = time.time()
    result1 = solve_n_queens_no_pruning(n)
    time1 = time.time() - start
    print(f"Thời gian: {time1:.6f}s")

    # Test có pruning
    print("\n[2] CÓ pruning:")
    start = time.time()
    result2 = solve_n_queens_with_pruning(n)
    time2 = time.time() - start
    print(f"Thời gian: {time2:.6f}s")

    # In kết quả so sánh
    if time2 > 0:
        print(f"\nTốc độ tăng: {time1/time2:.2f}x")
    else:
        print(f"\nTốc độ tăng: ∞ (pruning quá nhanh)")

    # In một giải pháp mẫu
    if len(result2) > 0:
        print(f"\nMột giải pháp cho {n}-Queens:")
        print_board(result2[0], n)


compare_n_queens(4)
compare_n_queens(6)
```

# Bài 3 – Subset Sum với pruning nâng cao

Cài đặt Subset Sum với nhiều kỹ thuật pruning.

## Phần A – Subset Sum cơ bản

```python
class SubsetCounter:
    """Class để đếm số lần gọi hàm và nhánh bị cắt"""
    def __init__(self):
        self.total_calls = 0
        self.pruned = 0
        self.solutions = 0

    def report(self, label=""):
        if label:
            print(f"\n--- {label} ---")
        print(f"Tổng số lần gọi: {self.total_calls}")
        print(f"Số nhánh đã cắt tỉa: {self.pruned}")
        print(f"Số giải pháp tìm được: {self.solutions}")
        total_considered = self.total_calls + self.pruned
        if total_considered > 0:
            print(f"Tỷ lệ cắt: {self.pruned / total_considered * 100:.2f}%")


def subset_sum_basic(nums, target):
    """
    Subset Sum cơ bản - KHÔNG có pruning
    Tìm tất cả tập con có tổng bằng target.
    """
    counter = SubsetCounter()
    result = []

    def backtrack(start, path, current_sum):
        counter.total_calls += 1

        # Base case: Nếu current_sum == target → lưu kết quả
        if current_sum == target:
            result.append(path.copy())
            counter.solutions += 1

        # Recursive case: Thử thêm các số từ vị trí start
        for i in range(start, len(nums)):
            # CHOOSE
            path.append(nums[i])

            # EXPLORE
            backtrack(i + 1, path, current_sum + nums[i])

            # UNCHOOSE
            path.pop()

    backtrack(0, [], 0)
    counter.report("Subset Sum cơ bản")
    return result, counter


# Test
print("=== Test Subset Sum cơ bản ===")
r1, _ = subset_sum_basic([2, 3, 6, 7], 7)
print(f"Tập con tổng = 7 từ [2,3,6,7]: {r1}")

r2, _ = subset_sum_basic([1, 2, 3, 4, 5], 5)
print(f"\nTập con tổng = 5 từ [1,2,3,4,5]: {r2}")
```

## Phần B – Subset Sum với pruning

Áp dụng 4 kỹ thuật pruning:
1. **Early termination:** Dừng nếu `current_sum > target`
2. **Sort và break:** Sắp xếp tăng dần, break nếu phần tử tiếp theo vượt target
3. **Skip duplicates:** Bỏ qua phần tử trùng lặp
4. **Remaining check:** Dừng nếu cộng hết cũng không đủ target

```python
def subset_sum_with_pruning(nums, target):
    """
    Subset Sum CÓ pruning nâng cao
    Áp dụng 4 kỹ thuật pruning.
    """
    counter = SubsetCounter()
    result = []

    # Kỹ thuật 2: Sắp xếp tăng dần để hỗ trợ pruning
    sorted_nums = sorted(nums)

    # Tính prefix sum từ cuối để hỗ trợ remaining check
    n = len(sorted_nums)
    remaining_sums = [0] * (n + 1)
    for i in range(n - 1, -1, -1):
        remaining_sums[i] = remaining_sums[i + 1] + sorted_nums[i]

    def backtrack(start, path, current_sum):
        counter.total_calls += 1

        # Base case: Tổng bằng target → lưu kết quả
        if current_sum == target:
            result.append(path.copy())
            counter.solutions += 1
            return

        # Kỹ thuật 1: Early termination – tổng đã vượt target
        if current_sum > target:
            counter.pruned += 1
            return

        # Recursive case: Thử thêm các số từ vị trí start
        for i in range(start, len(sorted_nums)):
            # Kỹ thuật 4: Remaining check – cộng hết cũng không đủ
            if current_sum + remaining_sums[i] < target:
                counter.pruned += 1
                break

            # Kỹ thuật 2: Sort và break – số tiếp theo làm vượt target
            if current_sum + sorted_nums[i] > target:
                counter.pruned += 1
                break  # Các số sau còn lớn hơn → break luôn

            # Kỹ thuật 3: Skip duplicates – bỏ qua số trùng
            if i > start and sorted_nums[i] == sorted_nums[i - 1]:
                counter.pruned += 1
                continue

            # CHOOSE
            path.append(sorted_nums[i])

            # EXPLORE
            backtrack(i + 1, path, current_sum + sorted_nums[i])

            # UNCHOOSE
            path.pop()

    backtrack(0, [], 0)
    counter.report("Subset Sum với pruning")
    return result, counter


# Test
print("=== Test Subset Sum với pruning ===")
r3, _ = subset_sum_with_pruning([2, 3, 6, 7], 7)
print(f"Tập con tổng = 7 từ [2,3,6,7]: {r3}")

r4, _ = subset_sum_with_pruning([1, 2, 3, 4, 5], 5)
print(f"\nTập con tổng = 5 từ [1,2,3,4,5]: {r4}")
```

## Phần C – So sánh và đo lường

Test với input lớn: `nums = list(range(1, 20))`, `target = 50`

```python
def compare_subset_sum(nums, target):
    print(f"\n{'='*60}")
    print(f"So sánh Subset Sum")
    print(f"nums = {nums[:10]}{'...' if len(nums) > 10 else ''} (n={len(nums)})")
    print(f"target = {target}")
    print(f"{'='*60}")

    # Test không có pruning
    print("\n[1] KHÔNG có pruning:")
    start = time.time()
    result1, counter1 = subset_sum_basic(nums, target)
    time1 = time.time() - start
    print(f"Thời gian: {time1:.6f}s")

    # Test có pruning
    print("\n[2] CÓ pruning:")
    start = time.time()
    result2, counter2 = subset_sum_with_pruning(nums, target)
    time2 = time.time() - start
    print(f"Thời gian: {time2:.6f}s")

    # Báo cáo so sánh
    print(f"\n{'─'*60}")
    print(f"{'BÁO CÁO SO SÁNH':^60}")
    print(f"{'─'*60}")
    print(f"{'Tiêu chí':<30} {'Không pruning':>12} {'Có pruning':>12}")
    print(f"{'─'*60}")
    print(f"{'Tổng lần gọi hàm':<30} {counter1.total_calls:>12} {counter2.total_calls:>12}")
    print(f"{'Số nhánh đã cắt':<30} {counter1.pruned:>12} {counter2.pruned:>12}")

    total1 = counter1.total_calls + counter1.pruned
    total2 = counter2.total_calls + counter2.pruned
    pct1 = (counter1.pruned / total1 * 100) if total1 > 0 else 0
    pct2 = (counter2.pruned / total2 * 100) if total2 > 0 else 0
    print(f"{'Tỷ lệ cắt (%)':<30} {pct1:>11.2f}% {pct2:>11.2f}%")
    print(f"{'Thời gian (s)':<30} {time1:>12.6f} {time2:>12.6f}")

    if time2 > 0:
        print(f"{'Tốc độ tăng':<30} {time1/time2:>11.2f}x")
    else:
        print(f"{'Tốc độ tăng':<30} {'∞':>12}")

    print(f"{'Số giải pháp':<30} {len(result1):>12} {len(result2):>12}")
    print(f"{'─'*60}")

    # In một vài giải pháp mẫu
    if result2:
        print(f"\nMột số giải pháp mẫu (tối đa 5):")
        for i, sol in enumerate(result2[:5]):
            print(f"  {i+1}. {sol} → tổng = {sum(sol)}")


compare_subset_sum(list(range(1, 20)), 50)
```
